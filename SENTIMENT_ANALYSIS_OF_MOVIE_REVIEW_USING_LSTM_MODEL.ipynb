### Importing Deep Learning Libraries

```python
# Import neural network tools from keras
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

print("Deep learning tools loaded successfully!")
```

### : Preparing Text Data (Movie Reviews)

```python
# Sample dataset of 6 movie reviews
reviews = [
    "This movie was amazing and fantastic",
    "I loved the story and great acting",
    "Beautiful masterpiece completely enjoyed it",
    "Worst movie ever very boring story",
    "Waste of time and terrible acting",
    "I hated this film it was awful"
]

# Labels: 1 = Positive, 0 = Negative
labels = np.array([1, 1, 1, 0, 0, 0])

print("Sample movie reviews dataset is ready!")
```

### Text Preprocessing and Word Tokenization

```python
# 1. Convert words into unique numbers
my_tokenizer = Tokenizer(num_words=100)
my_tokenizer.fit_on_texts(reviews)
sequences = my_tokenizer.texts_to_sequences(reviews)

# 2. Make all review lengths equal (Padding)
padded_reviews = pad_sequences(sequences, maxlen=6)

print("Processed text data into numbers:")
print(padded_reviews)
```

###  Building and Training the LSTM Neural Network

```python
# Create a blank sequential network model
my_lstm_model = Sequential()

# Add layers as required by the assignment
my_lstm_model.add(Embedding(input_dim=100, output_dim=8, input_length=6))
my_lstm_model.add(LSTM(units=8))
my_lstm_model.add(Dense(units=1, activation='sigmoid')) # Output layer for binary classification

# Compile the model
my_lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model for 10 epochs
print("Starting training loop...")
my_lstm_model.fit(padded_reviews, labels, epochs=100, verbose=1)
print("LSTM model training completed successfully!")
```

### Model Prediction on a New Review

```python
# Provide a new review to test the model
new_review = ["The film was fantastic and great"]

# Preprocess the new review exactly like before
new_seq = my_tokenizer.texts_to_sequences(new_review)
new_padded = pad_sequences(new_seq, maxlen=6)

# Make prediction
prediction_score = my_lstm_model.predict(new_padded)

# Print final readable result based on threshold
if prediction_score > 0.5:
    print(f"Review: '{new_review}' -> Prediction: POSITIVE REVIEW")
else:
    print(f"Review: '{new_review}' -> Prediction: NEGATIVE REVIEW")
```
