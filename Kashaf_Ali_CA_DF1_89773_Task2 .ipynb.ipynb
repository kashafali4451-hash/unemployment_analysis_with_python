### Step 1: Importing Libraries

```
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#step 2: loading the database


```
data = pd.read_csv('/content/Unemployment in India.csv')
```

#step 3: data cleaning and formatting

```
data.columns = data.columns.str.strip()
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
data = data.dropna()
```

### Step 4: Feature Extraction and Data Verification

```
data['Month'] = data['Date'].dt.strftime('%B')
data.head()
```

### Step 5: COVID-19 Impact Analysis using Line

```
plt.figure(figsize=(10, 5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data, color='blue')
plt.title('Unemployment Rate Timeline')
plt.show()
```

### Step 6: Seasonal Trends Analysis (Monthly Bar Chart)

```
# list of months in correct calendar order
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# simple bar plot for monthly analysis
plt.figure(figsize=(10, 5))
sns.barplot(x='Month', y='Estimated Unemployment Rate (%)', data=data, order=months_order)
plt.title('Monthly Unemployment Trends')
plt.xticks(rotation=30)
plt.show()
```

### Step 7: Regional Analysis for Economic Policies (State-wise Chart)

```
# Har state ka average nikal rahe hain
state_avg = data.groupby('Region')['Estimated Unemployment Rate (%)'].mean().reset_index()

# Horizontal bar plot bana rahe hain taake states ke naam saaf parhe jayein
plt.figure(figsize=(10, 8))
sns.barplot(x='Estimated Unemployment Rate (%)', y='Region', data=state_avg)
plt.title('State wise Average Unemployment Rate')
plt.show()
```
