### (Libraries)

```
# Importing everyday tools for our lab project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

print("All setup is ready!")
```

 ### Dataset Creation and Cleaning

```
# Simple house database with 10 rows to fix the nan error
my_data = {
    'Size': [1000, 1500, 1200, 1800, 2000, 1100, 1400, 1600, 2200, 1300],
    'Rooms': [2, 3, 2, 4, 4, 2, 3, 3, 5, 2],
    'Price': [50, 75, 60, 90, 100, 55, 70, 80, 110, 65]
}

# Changing dictionary to a dataframe table
df = pd.DataFrame(my_data)

# Preprocessing: checking for any empty cells or null data
print("Checking for missing data:")
print(df.isnull().sum())
```

###Exploratory Data Analysis (EDA)

```
# Plotting a basic scatter chart of Size vs Price
plt.scatter(df['Size'], df['Price'], color='red')
plt.title("House Size and Price Graph")
plt.xlabel("Size")
plt.ylabel("Price")
plt.show()
```

###Splitting Data into Train and Test Sets

```
# Inputs (X) and Output target (y)
X = df[['Size', 'Rooms']]
y = df['Price']

# Split into 80% train and 20% test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print("Data split successfully!")
```

###Training the 3 Machine Learning Models

```
# Model 1: Simple Linear Regression
model1 = LinearRegression()
model1.fit(X_train, y_train)

# Model 2: Decision Tree Model
model2 = DecisionTreeRegressor()
model2.fit(X_train, y_train)

# Model 3: Random Forest Model
model3 = RandomForestRegressor()
model3.fit(X_train, y_train)

print("All three separate models are trained now.")
```

###  Comparison and Student Conclusion

```
# Checking results on test data
score1 = r2_score(y_test, model1.predict(X_test))
score2 = r2_score(y_test, model2.predict(X_test))
score3 = r2_score(y_test, model3.predict(X_test))

print("Linear Regression Score:", score1 * 100)
print("Decision Tree Score:", score2 * 100)
print("Random Forest Score:", score3 * 100)

# Student Observation and Conclusion:
# I observed that all models give different output scores on test data.
# In conclusion, Linear Regression gives the most straightforward predictions for this small dataset.
```
